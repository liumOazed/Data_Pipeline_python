```python
df= []
csv_files = [x for x in os.listdir(DATA_DIR) if x.endswith(".csv")]
for filename in csv_files:
    #print(filename)
    year = filename.replace(".csv", '')
    csv_path = os.path.join(DATA_DIR,filename)
    this_df = pd.read_csv(csv_path)
    this_df['filename'] = filename
    this_df['Year'] = year
    df.append(this_df)
    #print(this_df.head(n=1))
```

## DATA CLEANING

```python
df2['Rank'] = -1
df2.head()
```

```python
df2 = df2.loc[:, ~df2.columns.str.contains('^Unnamed')]
```

```python
df2.head()
```

```python
df2['Domestic %'] = df2['%']
df2['Foreign %'] = df2['%.1']
```

```python
df2.head()
```

```python
df2 = df2.drop(columns=['%','%.1'])
```

```python
df2.head()
```

```python
to_clean = ['Worldwide', 'Domestic', 'Foreign']
 
def str_to_int(current_val):
    cleaned = current_val.replace('$', '').replace(',', '')
    try:
        cleaned = int(cleaned)
    except:
        cleaned = 0
    return cleaned

def clean_col(row):
    #print(row)
    for columns in to_clean:
        current_val = row[columns]
        row[columns] = str_to_int(current_val)
    return row

df_cleaned = df2.apply(clean_col, axis=1)
```

```python
df_cleaned.head()
```

```python
df_cleaned.dtypes
```

```python
df_cleaned.sort_values(by = ['Worldwide'], inplace=True, ascending =False)
df_cleaned.head()
```

```python
df_cleaned.reset_index(inplace=True, drop=True)
```

```python
df_cleaned.head()
```

```python
df_cleaned['Rank'] = df_cleaned.index + 1
```

```python
df_cleaned.head()
```

```python
datasets = os.path.join(CACHE_DIR, 'movies_all_cleaned.csv')
df_cleaned.to_csv(datasets, index= False)
```
